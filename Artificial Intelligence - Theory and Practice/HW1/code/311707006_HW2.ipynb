```python
import numpy as np
import matplotlib.pyplot as plt


def longcall(ST, K, call):  # ST 標的現價 K 執行價 call 權利金
    return np.maximum(ST-K, 0) - call


def shortcall(ST, K, call):
    return -(np.maximum(ST-K, 0) - call)


def longput(ST, K, put):
    return np.maximum(K-ST, 0) - put


def shortput(ST, K, put):
    return -(np.maximum(K-ST, 0) - put)
```
